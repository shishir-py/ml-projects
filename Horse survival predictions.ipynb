```python
import pandas as pd
import numpy as np
path="datasets/horse.csv"
df=pd.read_csv(path)
from IPython.display import display
pd.set_option("display.max_columns", None)
display(df)
import warnings
warnings.filterwarnings('ignore')
```

```python
# df.drop('hospital_number', axis=1, inplace=True)
df=df.copy()
```

```python
y=df['outcome']
X=df.drop('outcome',axis=1)
df=X.copy()
```

```python
df.columns
```

```python
cat_features=[]
num_features=[]
for col in df.columns:
    if df[col].dtype=="object":
        cat_features.append(col)
    else:
        num_features.append(col)
num_features
```

```python
binary_features=[

]
ordinal_features=[

]
nominal_features=[

]
for col in cat_features:
    if df[col].nunique() <=2:
        binary_features.append(col)
    elif df[col].dtype== "object":
        nominal_features.append(col)
    else:
        ordinal_features.append(col)
for cols in df.columns:
    if cols in nominal_features:
        most_frequent = df[cols].value_counts().index[0]
        df[cols].fillna(most_frequent, inplace=True)
    elif cols in ordinal_features:
        df[cols]=df[cols].fillna(df[cols].mean())
    else:
        pass
```

```python
print("Binary",binary_features)
print("0rdinal",ordinal_features)
print("nominal",nominal_features)
```

```python
def binary_encode(df,columns,encoder):
    df=df.copy()
    for cols in columns:
        df[cols] = encoder.fit_transform(df[cols])
    return df
def ordinal_encode(df,columns):
    df=df.copy()
    df= pd.get_dummies(df, columns=columns, prefix=ordinal_features)
    return df
def nominal_encode(df,columns):
    df=df.copy()
    df= pd.get_dummies(df, columns=columns, prefix=nominal_features)
    return df
def num_encode(df,columns):
    df = df.copy()
    for col in columns:
        df[col].fillna(df[col].median(), inplace=True)
    return df

```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lb=LabelEncoder()
df_encoded = df.copy()
df_encoded = binary_encode(df_encoded, binary_features, lb)
df_encoded = ordinal_encode(df_encoded, ordinal_features)
df_encoded = nominal_encode(df_encoded, nominal_features)
df_encoded = num_encode(df_encoded, num_features)
```

```python
mapping={'lived':1,'died':0,'euthanized':2}
y=y.replace(mapping)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
pca=PCA()
pca.fit(df_encoded)
df_pca=pca.fit_transform(df_encoded)
scaled_X=sc.fit_transform(df_pca)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df_pca,y,train_size=0.8,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error,f1_score,recall_score
dc=DecisionTreeClassifier()
lr=LogisticRegression()
rf=RandomForestClassifier()
dc.fit(X_train,y_train)
```

```python
preds=dc.predict(X_test)
```

```python
accuracy=mean_absolute_error(preds,y_test)
```

```python
def model_select(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    acc=mean_absolute_error(preds,y_test)
    return acc

model_select(lr,X_train,y_train,X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Define the param grid
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'penalty': [ 'l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'l1_ratio': [0, 0.5, 1]
}

# Create a logistic regression object
logistic_regression = LogisticRegression()

# Create the GridSearchCV object
grid_search = GridSearchCV(logistic_regression, param_grid, cv=5)

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```
